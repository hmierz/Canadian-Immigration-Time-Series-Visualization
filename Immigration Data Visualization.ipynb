```python
#Context: India and Pakistan, once part of the same country, have had different political and
#economic trajectories since their partition in 1947.
#Your Task: Create two histograms side by side, one for Indian immigration and one for
#Pakistani immigration to Canada from 1980 to 2013. How do these distributions compare?
#What might account for any differences you observe in your histograms?
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example DataFrame; replace with your actual DataFrame
# canada_data_indexed = pd.read_csv('path_to_your_file.csv', index_col=0)

# Filter the DataFrame to get data for India and Pakistan
india_data = canada_data_indexed.loc['India']
pakistan_data = canada_data_indexed.loc['Pakistan']

# Ensure we only consider the years 1980 to 2013
india_data = india_data.loc['1980':'2013']
pakistan_data = pakistan_data.loc['1980':'2013']

# Set Seaborn theme for nicer styling
sns.set_theme(style="whitegrid")

# Create a figure with two subplots for side-by-side histograms
fig, axs = plt.subplots(1, 2, figsize=(14, 6), sharey=True)  # Share y-axis for easier comparison

# Plotting histogram for Indian immigration using Seaborn
sns.histplot(india_data, bins=10, kde=True, color="skyblue", edgecolor='black', ax=axs[0])
axs[0].set_title('Indian Immigration to Canada (1980-2013)')
axs[0].set_xlabel('Number of Immigrants')
axs[0].set_ylabel('Number of Years')

# Plotting histogram for Pakistani immigration using Seaborn
sns.histplot(pakistan_data, bins=10, kde=True, color="lightgreen", edgecolor='black', ax=axs[1])
axs[1].set_title('Pakistani Immigration to Canada (1980-2013)')
axs[1].set_xlabel('Number of Immigrants')

# Adjust layout to avoid overlap
plt.tight_layout()
plt.show()
```

```python
#These distributions compare in that India's X axis begins at a much higher number of immigrants than Pakistan's. They both have a high outlier showing peaks in immigration, and India has a very low dip in immigration as well. Pakistan's dips remained around the same value each time. Both KDEs trend downwards at the end, and have similar shapes overall beside's India's peeking a bit higher. This could be due to India's much larger population size. Ther are also significant differences between India and Pakistan in terms of economics, quality of life, and ability to emigrate.
```

```python
#2.4.1 Comparing Immigration Variability: Japan vs. South Korea
#Context: Japan and South Korea are both highly developed East Asian economies but have
#different demographic challenges.
#Your Task: Create box plots for Japanese and South Korean immigration to Canada from
#1980 to 2013. How does the variability in immigration numbers compare between these two
#countries? What factors might contribute to any differences you observe in your box plots?
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example DataFrame; replace with your actual DataFrame
# canada_data_indexed = pd.read_csv('path_to_your_file.csv', index_col=0)

# Filter the DataFrame to get data for Japan and Republic of Korea
japan_data = canada_data_indexed.loc['Japan']
korea_data = canada_data_indexed.loc['Republic of Korea']  # Data under 'South Korea'

# Ensure we only consider the years 1980 to 2013
japan_data = japan_data.loc['1980':'2013']
korea_data = korea_data.loc['1980':'2013']

# Prepare the data for plotting in a single DataFrame
df = pd.DataFrame({
    'Year': japan_data.index,
    'Japan': japan_data.values,
    'Republic of Korea': korea_data.values  # Updated label
})

# Reshape the DataFrame for Seaborn's boxplot function
df_melted = df.melt(id_vars='Year', value_vars=['Japan', 'Republic of Korea'], var_name='Country', value_name='Immigrants')

# Set Seaborn theme
sns.set_theme(style="whitegrid")

# Create the box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Country', y='Immigrants', data=df_melted, palette="Set2")

# Add title and labels
plt.title('Immigration to Canada (1980-2013): Japan vs Republic of Korea')
plt.xlabel('Country')
plt.ylabel('Number of Immigrants')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#South Korea has way higher immigration numbers than Japan, probably due to a larger population size. Both countries also have outliers. This could be due to the more recent development of South Korea's economy compared to Japan's, as well as cultural attitudes such as Japan being a highly collectivistic culture.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example DataFrame; replace with your actual DataFrame
# canada_data_indexed = pd.read_csv('path_to_your_file.csv', index_col=0)

# Define the countries for which we want to create box plots
countries = ['Russian Federation', 'India', 'China', 'South Africa']

# Set Seaborn theme
sns.set_theme(style="whitegrid")

# Create a figure with subplots
plt.figure(figsize=(16, 10))

for i, country in enumerate(countries, start=1):
    # Filter the DataFrame to get data for the current country
    country_data = canada_data_indexed.loc[country]

    # Ensure we only consider the years 1980 to 2013
    country_data = country_data.loc['1980':'2013']

    # Reset index to convert years to a column
    df_country = country_data.reset_index()
    df_country.columns = ['Year', 'Immigrants']

    # Convert Year to string to ensure consistency
    df_country['Year'] = df_country['Year'].astype(str)

    # Create a subplot for each country
    plt.subplot(2, 2, i)  # 2 rows, 2 columns, i-th subplot
    sns.boxplot(y='Immigrants', data=df_country, color="skyblue", width=0.5)
    plt.title(f'{country} Immigration to Canada (1980-2013)')
    plt.ylabel('Number of Immigrants')
    plt.xlabel('')

# Adjust layout
plt.tight_layout()
plt.show()
```
